## 3: Data Analysis, Preparation and Wrangling

### Handling missing rows

```python
df.isna() #Presence of True indicates there is missing data. False indicates there is no missing data or NAN
```

### Handling the currency symbols and comma format

```python
#removing the currency symbol and the comma character

#replacing the string currency symbols with empty space. 
df.price = df.price.apply(lambda x: str.replace(x,'£', ''))

#replacing the comma character with empty space.
df.price = df.price.apply(lambda x: str.replace(x,',', ''))
```

## 5: Simple Linear Regression Model

```python
#(Target varible) assigned to Y variable
Y = df2.price 

#Features assigned to X variable
X = df2.drop('price',axis = 1)  
X
```

```python
from sklearn.model_selection import train_test_split
#Splitting dataset with a 25/75 ratio
X_train, X_test, Y_train, Y_test = train_test_split(X, 
                                                  Y,
                                                   test_size=0.25,
                                                   random_state = 42)
print("Training and testing split was successful.")
```

```python
#Building the model. 
# Import the LinearRegression class from the sklearn library
from sklearn.linear_model import LinearRegression

#Initialize a Linear Regression model
model = LinearRegression()

#Fit the model to the training data
model.fit(X_train,Y_train)
```

```python
#Getting the predictions for the test dataset
predictions = model.predict(X_test)
```

```python
#Calculating the performance metrics
r2 = model.score(X_test,Y_test)
print ("R^2 is: \n", r2)
```

```python
#Import the mean_squared_error function from the sklearn library
from sklearn.metrics import mean_squared_error
#Calculate the root mean squared error (RMSE)

rmse = mean_squared_error(Y_test, predictions)**0.5
#print out
print ('RMSE is: \n', rmse)
```

```python
#Assign the actual target values to a variable
actual_values = Y_test

#Create a scatter plot of the predicted values against the actual values
#The 'alpha' argument controls the transparency of the points, with a value of 0 being completely transparent and 1 being completely opaque
#The 'color' argument sets the color of the points in the scatter plot, in this case red (r)

plt.scatter(predictions, actual_values, alpha=.7,
            color='r') 

#Label the x-axis as 'Predicted Price'
plt.xlabel('Predicted Price')

#Label the y-axis as 'Actual Price'
plt.ylabel('Actual Price')

#Give the plot a title of 'Linear Regression Model'
plt.title('Linear Regression Model')

#Set y-axis to start at 0
plt.ylim(0, max(actual_values) + 10)  # I added 10 as a buffer; you can set it to whatever you like


# Adding regression line
slope, intercept = np.polyfit(predictions, actual_values, 1)  # Fit a 1st degree polynomial (a line) to the data
x_values = np.linspace(min(predictions), max(predictions), 100)  # Generate 100 x-values from the minimum to the maximum predicted value
y_values = slope * x_values + intercept  # Calculate the corresponding y-values based on the slope and intercept


plt.plot(x_values, y_values, color='blue')  # Add the regression line


# Display the scatter plot
plt.show()
```

```python
#Get the names of the features used in the training data
feats = list(X_train.columns)

#Get the coefficients of the linear regression model
coef = model.coef_.transpose()

#Combine the feature names and coefficients into a 2D array
coef_table = np.vstack((feats, coef)).T

#Convert the 2-D array into a pandas DataFrame
df_coef = pd.DataFrame(coef_table, columns = ['Feature', 'Coefficients'])

#Display the DF
df_coef
```

## 6: Linear Regression Model with Lasso/Ridge


```python
#Import the Ridge and Lasso classes from the sklearn library
from sklearn.linear_model import Ridge, Lasso
```

```python
# Split the data into target"Y1" and features "X1"
Y1 = df1['price']
X1 = df1.drop('price', axis=1)
```

```python
#Split the data into training and testing sets with 25/75 ratio
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, 
                                                  Y1,
                                                   test_size=0.25,
                                                   random_state = 42)
```

```python
#Create an array of alpha angles to be tested
alphas = np.linspace(1, 0.01, 10)
alphas
```

```python
#Initialize lists to store the RMSE and R2 values for each alpha
rmse_list = []
r2_list = []

#loop through each alpha value
for a in alphas:
    #Run the model and evaluate for each alpha value
    lasso = Lasso(alpha = a, max_iter=10000)
    lasso.fit(X1_train,Y1_train)
    pred = lasso.predict(X1_test)
    
    r2_list.append(lasso.score(X1_test,Y1_test))
    rmse = mean_squared_error(Y1_test, pred)**0.5;
    rmse_list.append(rmse)
    
#Stack the alpha values, RMSE values, and R2 values into a 2-D array    
lasso_result = np.vstack((alphas, rmse_list, r2_list)).T

#Convert the 2-D array into a pandas DataFrame
lasso_df = pd.DataFrame(lasso_result, columns=['Alpha', 'RMSE', 'R2'])
lasso_df
```

```python
#The best alpha value was 0.67 that gave R2 of 0.756922 and RMSE of 4635.390617	 
lasso = Lasso(alpha = 0.67)

#Fit the model to training data
lasso.fit(X1_train,Y1_train)

#Use the trained model to make predictions on the test data X1_test
pred = lasso.predict(X1_test)
```

```python
#Printing r^2 value and RMSE
print("R2:", lasso.score(X1_test, Y1_test))
print("RMSE:", mean_squared_error(Y1_test, pred)**0.5)


print(pd.Series(lasso.coef_, index=X1_test.columns))
```

```python
#Plot the predicted values against the actual test values Y1_test
x_ax = range(len(pred))
plt.scatter(x_ax, Y1_test, s=5, color="blue", label="original")
plt.plot(x_ax, pred, lw=0.8, color="red", label="predicted")

#Add a legend to the plot
plt.legend()

#Show the plot
plt.show()
```

## 7: Polynomial Regression Model



```python
#Importing the PolynomialFeatures library to perform polynomial regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Initializing lists to store the results of the polynomial regression model for different degrees
pr2_list = []
prmse_list = []
pdegrees_list = []


for i in range(1,6):
    ply = PolynomialFeatures(degree=i) #Initializing a PolynomialFeatures object with degree=i
    plyfeatures = ply.fit_transform(X1) #Transforming the features using the polynomial features object
    xtrain, xtest, ytrain, ytest = train_test_split(plyfeatures, Y1, test_size=0.25, random_state=42)

    Ply_linear = LinearRegression() #Initialize the model
    Ply_linear.fit(xtrain, ytrain) #Fit the training sets to the model
    pred1 = Ply_linear.predict(xtest) #Predict the price using test data
    pr2_list.append(Ply_linear.score(xtest, ytest)) #Append the R^2 score
    prmse_list.append(mean_squared_error(ytest, pred1)**0.5) #Append the RMSE 
    pdegrees_list.append(i) #Appending the degree

    
#stack the results in a table
ply_table = pd.DataFrame(np.vstack((pdegrees_list, pr2_list, prmse_list)).T, columns=['Degree','R^2', 'RMSE'])

ply_table


```

From the results the degree that gives us the best R^2 is degree 3. 

```python
#Fitting a polynomial regression model with degree=3
ply = PolynomialFeatures(degree=3)
plyfeatures = ply.fit_transform(X1)
xtrain, xtest, ytrain, ytest = train_test_split(plyfeatures, Y1, test_size=0.25, random_state=42)
Ply_linear = LinearRegression()
Ply_linear.fit(xtrain, ytrain)
prediction1 = Ply_linear.predict(xtest)
    
```

```python
#Checking R^2 value of the model
Ply_linear.score(xtest, ytest)
```

```python
#Checking RMSE value of the model
mean_squared_error(ytest,prediction1)**0.5
```

```python
#Creating scatter plot
a_ax = range(len(prediction1))
plt.scatter(x_ax, ytest,s=5, color="blue", label="original")
plt.plot(x_ax, prediction1, lw=0.8, color='red', label="predicted")
plt.legend()
plt.show()
```

### Polynomial Regression Model with Lasso

```python
#Define the array of alphas from 0 to 1
PlyLasso_Alphas = np.linspace(0,1, 10)


#Initialize lists to store R^2 and RMSE values for each alpha
PlyLasso_r2 = []
PlyLasso_rmse = []

#Create an instance of the Lasso model
Ply_lasso = Lasso()

#Loop through the alpha values
for a in PlyLasso_Alphas:
    Ply_lasso.set_params(alpha=a)
    Ply_lasso.fit(xtrain, ytrain)
    pred1Ply_lasso = Ply_lasso.predict(xtest)
    PlyLasso_r2.append(Ply_lasso.score(xtest,ytest))
    PlyLasso_rmse.append(mean_squared_error(ytest, pred1Ply_lasso)**0.5)

#Create a DataFrame to store the alpha, R^2 and RMSE values 
ply_lasso_df = pd.DataFrame(np.vstack((PlyLasso_Alphas, PlyLasso_r2 , PlyLasso_rmse)).T, 
                                     columns=['Alpha', 'R^2', 'RMSE'])
#print df    
ply_lasso_df
```
